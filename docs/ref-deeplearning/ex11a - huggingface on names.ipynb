```python
# import
import numpy as np
import tensorflow as tf
from tqdm import tqdm
```

```python
# define names
names = [
    ['George', 'Washington'],
    ['John', 'Adams'],
    ['Thomas', 'Jefferson'],
    ['James', 'Madison'],
    ['James', 'Monroe'],
    ['John', 'Adams'],
    ['Andrew', 'Jackson']]
names = np.asarray(names)
names
```

```python
# https://huggingface.co/bert-base-uncased
# import
try:
    from transformers import BertTokenizer, TFBertModel
except:
    !pip install transformers
    from transformers import BertTokenizer, TFBertModel
```

```python
# load models
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
model = TFBertModel.from_pretrained("bert-base-uncased")
```

```python
# get model
def bert_encoder_(text):
    """
    The function `bert_encoder_` encodes a string of text.
    text: string
    output: float that is the sum of the output layer of the bert model using the text as input
    """
    encoded_input = tokenizer(text, return_tensors='tf')
    output = model(encoded_input)
    return np.asarray(output[0]).sum()
```

```python
# test a word
bert_encoder_('hello')
```

```python
# test a name
i=1
[names[i,0], names[i,1], bert_encoder_(names[i,0]), bert_encoder_(names[i,1])]
```

```python
%%time 
results_ = []
for i in tqdm(range(names.shape[0])):
    results_.append([names[i,0], names[i,1], bert_encoder_(names[i,0]), bert_encoder_(names[i,1])])

# convert to array from list
results_ = np.asarray(results_)
```
