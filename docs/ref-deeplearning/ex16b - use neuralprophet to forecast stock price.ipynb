## Library

Import the libraries.

````python
import yfinance as yf
import pandas as pd
from neuralprophet import NeuralProphet
from neuralprophet.utils import set_random_seed
````

Download the historical prices of the S&P500 index and prepare the DataFrame for modeling with NeuralProphet:

````python
df = yf.download("HUBS") #$, start="2010-01-01", end="2023-11-22")
df = df[["Adj Close"]].reset_index(drop=False)
df.columns = ["ds", "y"]
````

Create train/test split:

````python
TEST_LENGTH = 60
df_train = df.iloc[:-TEST_LENGTH]
df_test = df.iloc[-TEST_LENGTH:]
````

````python
df_train.head(3)
````

````python
df_train.shape, df_test.shape
````

## Basic Structure

This is something quick and dirty using default parameters. The code fits the data and makes a prediction in sample all at once.

```py
# step 1. define model
m = NeuralProphet(
    ar_reg=1
)
m.set_plotting_backend("plotly")

# step 2. fit the data
metrics = m.fit(df_train)

# step 3. in-sample prediction and plot it
forecast = m.predict(df_train)
m.plot(forecast)

# step 4. create future data and predict
future = m.make_future_dataframe(df, periods=30*2, n_historic_predictions=len(df))
forecast = m.predict(future)
m.plot(forecast.tail(30*5))
```

### Step 1: Define the Model
- We initiate a `NeuralProphet` model with the autoregression regularization (`ar_reg`) set to 1.
- The plotting backend is set to "plotly" for enhanced interactive plots.

### Step 2: Fit the Model
- The model is fitted to the training data `df_train` and the metrics of the training process are stored.

### Step 3: In-sample Prediction and Plot
- We perform in-sample predictions on the training dataset to evaluate how well the model has learned the underlying patterns.
- The predictions are then plotted using the model's built-in `plot` method with Plotly.

### Step 4: Future Data Prediction and Plotting
- We create a dataframe for future dates using the `make_future_dataframe` method with 60 future days and historical data.
- The model makes predictions for these future dates and the last 150 days of the forecast are plotted to visualize the model's future predictions.

## Usage

Run the Python script to train the model and view the forecasts. Ensure you have the `NeuralProphet` library installed along with `plotly` for plotting.


### Model Fitting

````python
# Model and prediction
m = NeuralProphet(
    changepoints_range=0.95,
    ar_reg=2,
    # daily_seasonality=True,
    yearly_seasonality=True
)
m.set_plotting_backend("plotly")
metrics = m.fit(df_train, epochs=70)
forecast = m.predict(df_train)
m.plot(forecast)
````

### Create Future Data

````python
future = m.make_future_dataframe(df, periods=30*2, n_historic_predictions=len(df))
forecast = m.predict(future)
````

### Plot Future Prediction

````python
m.plot(forecast.tail(365))
````

## Building and Improving Models

From this section and on, we keep building complexity of our NeuralProphet model to increase the test set performance.

````python
set_random_seed(42)
model = NeuralProphet(changepoints_range=0.95)
metrics = model.fit(df_train, freq="D") # set freq to "D" as daily data
````

The following plots the training performance. You want to see the errors going down as `epoch` goes up.

````python
(
    metrics.drop(columns=["RegLoss"]).plot(
        title="Evaluation metrics during training",
        subplots = True
    )
)
````

Calculuate the predictions and plot the fit.

````python
pred_df = model.predict(df)
pred_df.shape
````

### In-sample Prediction Plot

The following code plots the training performance using in-sample data. Most of the time the prediction `yhat` should follow the data `y`.

````python
pred_df.plot(
    x="ds",
    y=["y", "yhat1"],
    title="S&P 500 - forecast vs ground truth"
)
````

The following produces out-of-sample (or also known as test set) performance. As indicated using `-TEST_LENGTH` the test data is defined in the beginning of the code.

In this case, since this is out-of-sample data, it is unlikely the first time you will see the prediction aligns with the data. You'll have to go back up to tune the parameters of the `neuralprophet` model.

````python
(
    pred_df.iloc[-TEST_LENGTH:].plot(
        x="ds",
        y=["y", "yhat1"],
        title="S&P 500 - forecast vs ground truth"
    )
)
````

### Add AR Terms

AR components are not included before. Recall

```py
# step 1. define model
m = NeuralProphet(
    ar_reg=1
)
m.set_plotting_backend("plotly")

# step 2. fit the data

# step 3. in-sample prediction and plot it

```

Add the AR components to NeuralProphet:

````python
set_random_seed(42)
model = NeuralProphet(
    # changepoints_range=0.95,
    n_lags=10,
    ar_reg=1,
)

metrics = model.fit(df_train, freq="D")

pred_df = model.predict(df)
model.plot(pred_df)
````

The following plots the plot for out-of-sample performance. Usually for stock price prediction, `AR` terms have higher chance of bringing the prediction `yhat` to the data and make them align.

Simply put, if you see the two lines below are close to each other, it means the model does a good job at prediction on test set.

````python
# pred_df.iloc[-TEST_LENGTH:].plot(
#     x="ds",
#     y=["y", "yhat1"],
#     title="S&P500 - forecast vs ground truth")
````

````python
future = model.make_future_dataframe(df, periods=365, n_historic_predictions=len(df))
print(future.tail())
forecast = model.predict(future)
model.plot(forecast.tail(30*3))
````
